```python
# ------ UTILIZANDO A BASE movies.

import sqlite3
import pandas as pd

conn = sqlite3.connect('data/movie_database.db')
def consulta(query):
    return pd.read_sql(query, conn)
```

```python
# Quais tabelas existem no banco de dados?

res = conn.execute("SELECT name FROM sqlite_master WHERE type = 'table'")

res.fetchall()
```

```python
# Crie um dicionário de dados.

dados = {}

cursor = conn.cursor()

names = []

var_names = []

var_type = []

check_null = []

#Master listing of all database objects in the database and the SQL used to create each object.

cursor.execute('SELECT name FROM sqlite_master WHERE TYPE = "table"')

name = cursor.fetchall()

for n in name: # [("Directors",), ("Actors",) ..]
    nome = n[0]
    cursor.execute(f"PRAGMA table_info({nome})") #[(0,id,integral,0,0,1), (...), (...)]
    for row in cursor.fetchall(): 
        names.append(n[0])
        var_names.append(row[1])
        var_type.append(row[2])
        #SELECT COUNT(*) FROM your_table WHERE your_column IS NULL;
        cursor.execute(f'SELECT COUNT(*) FROM {nome} WHERE {row[1]} IS NULL;')
        check = cursor.fetchall() # returns list of tuples ex.[(0,)]
        if check[0][0] >0: 
            check_null.append("Yes")
        else: 
            check_null.append("No")




dados = {"Table" : names, "Variable Names": var_names , "Variables Types" : var_type, "Null Values": check_null}

df_music = pd.DataFrame(dados)

print(df_music)

```

```python
# Liste todos os filmes de drama.

consulta("""SELECT
            m.id,
            title MovieName,
            year Year,
            rating Rating,
            g.name Genre
            FROM movies m
                INNER JOIN genres g
                ON m.genre_id = g.id
            WHERE name = "Drama"
        """)
```

```python
# Existem quantos filmes de suspense?

consulta("""SELECT
            g.name Genre,
            COUNT(g.name) QtyGenre
            FROM movies m
                INNER JOIN genres g
                ON m.genre_id = g.id
            WHERE name = "Thriller"
        """)
```

```python
 # Apresente todos os filmes de Morgan Freeman

consulta("""SELECT
            m.id FilmID,
            m.title MovieName,
            m.year Year,
            m.rating Rating,
            a.name ActorName   
            FROM movies m
                INNER JOIN roles r
                ON m.id = r.movie_id
                INNER JOIN actors a
                ON a.id = r.actor_id
                WHERE a.name = "Morgan Freeman"
        """)
```
