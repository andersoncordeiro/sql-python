```python
import sqlite3
import pandas as pd

conn_movie  = sqlite3.connect('data/movie_database.db')
conn_music = sqlite3.connect('data/music-store.db')

def consulta(query, conn): 
    return pd.read_sql(query, conn)
```

```python
cursor_movie = conn_movie.cursor()
cursor_music = conn_music.cursor()
```

```python
query = "SELECT name FROM sqlite_master WHERE type='table';"
```

```python
tabelas_movies = pd.read_sql_query(query, conn_movie)
tabelas_music  = pd.read_sql_query(query, conn_music)
```

```python
#Quais tabelas existem no banco de dados?
print(tabelas_movies,f'\n {tabelas_music}')
```

```python
#Exiba todos os filmes da movies
df_movies = consulta("SELECT * FROM movies", conn_movie)
df_movies

```

```python
#Qual o filme mais novo?
df_movies_novo = consulta("SELECT id, title, rating , genre_id, director_id, MAX(year) AS year_new FROM movies ", conn_movie)
df_movies_novo 

```

```python
#E o mais antigo?
df_movies_velho = consulta("SELECT id, title, rating , genre_id, director_id, MIN(year) AS year_old FROM movies ", conn_movie)
df_movies_velho
```

```python
#Quais são os 5 filmes mais bem avaliados?
df_movies_rating_TOP = consulta("SELECT * FROM movies ORDER BY  rating DESC LIMIT 5" , conn_movie)
df_movies_rating_TOP
```

```python
#E os 10 piores filmes?
df_movies_rating_down = consulta("SELECT * FROM movies ORDER BY  rating ASC LIMIT 10" , conn_movie)
df_movies_rating_down
```

```python
#Liste todos os filmes de drama.
query = """
    SELECT  
    *
    FROM movies m 
        JOIN genres g 
        ON   m.genre_id = g.id
    WHERE g.name     = 'Drama'
"""
df_movies_drama = consulta( query, conn_movie)
df_movies_drama
```

```python
#Existem quantos filmes de suspense?
query = """
    SELECT  
    *
    FROM movies m 
        JOIN genres g 
        ON   m.genre_id = g.id
    WHERE g.name     = 'Suspense'
    
"""
df_movies_suspense = consulta( query, conn_movie)
df_movies_suspense
#0 filmes
```

```python
#Apresente todos os filmes de Morgan Freeman

df_movies_MF = consulta(("""
        SELECT
        m.id,
        m.title,
        m.year,
        m.rating,
        a.name
        FROM movies m 
            JOIN roles r 
                ON m.id = r.movie_id
            JOIN actors a
                ON r.actor_id = a.id
        WHERE a.name LIKE '%Freeman%'
            """), conn_movie)
df_movies_MF
```

```python
#Qual artista possui mais albuns? 
query = ("""
SELECT 
ar.name,
COUNT(al.artist_id) AS contagem 
	FROM albums al 
		JOIN artists ar 
		ON al.artist_id = ar.id
GROUP  BY artist_id 
ORDER BY contagem DESC
LIMIT 1
 """)
df_music_artist_album = consulta( query, conn_music)
df_music_artist_album

```

```python
#Existem clientes de quantos países? 
query = ("""
SELECT
c.country,
COUNT(DISTINCT c.country) AS contagem

	FROM customers c 
GROUP BY c.country
ORDER BY contagem DESC


 """)
df_music_customer_country = consulta( query, conn_music)
df_music_customer_country

#24 paises
```

```python
#Quantos clientes são brasileiros?
query = ("""
SELECT
c.country,
COUNT( c.country) AS contagem

	FROM customers c 
WHERE c.country = 'Brazil'
GROUP BY c.country
ORDER BY contagem DESC


 """)
df_music_customer_brazil = consulta( query, conn_music)
df_music_customer_brazil

```

```python
#Qual genero possui mais músicas?

query = ("""
SELECT
	ge.name,
    COUNT(tr.genre_id) AS contagem
FROM tracks tr 
    JOIN genres ge 
    ON tr.genre_id = ge.id 
WHERE 1 = 1
GROUP BY tr.genre_id
LIMIT 1
 """)
df_music_genre_music = consulta( query, conn_music)
df_music_genre_music
```

```python
#Qual genero possui mais albuns? 
query = ("""
SELECT
	ge.name,
    COUNT(tr.album_id) AS contagem
FROM tracks tr 
    JOIN genres ge 
    ON tr.genre_id = ge.id
    JOIN albums al
    ON tr.album_id = al.id 
WHERE 1 = 1
GROUP BY tr.album_id
ORDER BY contagem DESC
LIMIT 1
 """)
df_music_genre_album = consulta( query, conn_music)
df_music_genre_album
```

```python
#Exiba os 5 clientes que mais fizeram pedidos. Qual é o valor da maior compra?
query = ("""
SELECT 
	i.customer_id,
	c.first_name ||' '|| last_name as full_name,
	i.total ,
	COUNT(i.id) as pedidos
FROM invoices i
	JOIN customers c 
	ON i.customer_id = c.id 
GROUP BY i.customer_id 
ORDER BY i.total DESC
LIMIT 5

 """)
df_music_invoice_customer = consulta( query, conn_music)
df_music_invoice_customer
```

```python
#Quais os emails dos 10 clientes que mais gastaram na loja? 
query = ("""
SELECT 
	i.customer_id,
	c.first_name ||' '|| last_name as full_name,
    c.email,
	i.total ,
	COUNT(i.id) as pedidos
FROM invoices i
	JOIN customers c 
	ON i.customer_id = c.id 
GROUP BY i.customer_id 
ORDER BY i.total DESC
LIMIT 10

 """)
df_music_invoice_customer_mail = consulta( query, conn_music)
df_music_invoice_customer_mail
```

```python
#Qual o valor médio de cada compra? 

query = ("""
SELECT 

	AVG(i.total) as med_pedidos
FROM invoices i
	

 """)
df_music_invoice_avg = consulta( query, conn_music)
df_music_invoice_avg
```

```python
#Quantas compras estão acima da média?

query = ("""
SELECT
	i.customer_id,
	i.total 	
FROM invoices i
WHERE i.total > (SELECT 
				AVG(i.total) as med_pedidos
				FROM invoices i)

 """)
df_music_invoice_avg_up = consulta( query, conn_music)
df_music_invoice_avg_up['customer_id'].count
```

```python
# Vamos fazer uma playlist com as 12 músicas mais vendidas! 
# Quais foram elas? Quais os generos e os artistas?
query = ("""
SELECT 
	tr.name,
	ge.name,
	ar.name,
	COUNT(i.id) as contagem
FROM tracks tr 
	JOIN invoice_lines il 
		ON tr.id = il.track_id
	JOIN invoices i 
		ON il.invoice_id = i.id
	JOIN albums al 
		ON tr.album_id = al.id
	JOIN artists ar 
		ON al.artist_id  = ar.id
	JOIN genres ge 
		ON tr.genre_id = ge.id
GROUP BY tr.name
ORDER BY contagem DESC
LIMIT 12
""")
df_music_12_playlist = consulta( query, conn_music)
df_music_12_playlist

```

```python

# DESAFIO DO PROF: Precisamos definir 2 países para ações de marketing. 
# Para um, iremos direcionar campanhas dos albums mais caros, 
# nosso foco são os clientes que não se preocupam com o valor do pedido. 
# Qual seria sua estratégia?
#  Quais países poderíamos atender? 
# Quais albuns iremos recomendar para cada um dos países? 
# Apresente a query (ou as queries), a tabela (ou as tabelas) 
# e justifique suas escolhas.

query = ("""
SELECT 
	--i.customer_id,
	--c.first_name ||' '|| last_name as full_name,
	i.billing_country,
	SUM(i.total) as total_gasto,
	COUNT(i.id) as pedidos
FROM invoices i
GROUP BY i.billing_country 
ORDER BY total_gasto DESC
LIMIT 1
""")
df_music_gastador = consulta( query, conn_music)
df_music_gastador
```

```python
#ao analisar o preço médio do pedido , temos países com pouco pedido na liderança e a bélgica ainda é a mão de vaca
query = ("""
SELECT 
	--i.customer_id,
	--c.first_name ||' '|| last_name as full_name,
	i.billing_country,
	SUM(i.total) as total_gasto,
	COUNT(i.id) as pedidos,
	(SUM(i.total) / COUNT(i.id)) as total_pedido
FROM invoices i
GROUP BY i.billing_country 
ORDER BY total_pedido DESC 
""")
df_music_estudo = consulta( query, conn_music)
df_music_estudo
```

```python
query = ("""

SELECT 
	al.title,
	ar.name artist,
	ge.name genero,
	SUM(tr.unit_price) AS price_album
	--COUNT(i.id) as contagem
FROM tracks tr 
	left JOIN albums al 
		ON tr.album_id = al.id
	JOIN artists ar 
		ON al.artist_id  = ar.id
	JOIN genres ge 
		ON tr.genre_id = ge.id
GROUP BY al.title
ORDER BY price_album DESC  

""")
df_music_estudo_price = consulta( query, conn_music)
df_music_estudo_price
```
