## UTILIZANDO a BASE MUSIC_STORE


```python
import sqlite3
import pandas as pd

conn1 = sqlite3.connect('data/music-store.db')
def consulta1(query):
    return pd.read_sql(query, conn1)
```

### Qual artista possui mais albuns? 

```python
consulta1("""
            SELECT 
                artists.name, 
                COUNT(albums.artist_id) qtde_albums 
            FROM 
                albums 
                JOIN artists 
                    ON albums.artist_id = artists.id 
            GROUP BY 
                artists.name 
            ORDER BY 
                qtde_albums DESC
        """)

```

### Qual genero possui mais músicas?

```python
consulta1("""
            SELECT 
                genres.name, 
                COUNT(genre_id) qtde 
            FROM 
                tracks 
                JOIN genres 
                    ON tracks.genre_id = genres.id 
            GROUP by 
                genre_id 
            ORDER BY 
                genre_id 
            LIMIT 1
        """)
```

### Qual genero possui mais albuns? 

```python
consulta1("""
            SELECT 
                genres.name, 
                tracks.genre_id qtd, 
                COUNT(albums.id) 
            FROM 
                tracks 
                JOIN genres 
                    ON tracks.genre_id = genres.id 
                JOIN albums 
                    ON tracks.album_id = albums.id 
            GROUP BY 
                genres.name 
            ORDER BY 
                COUNT(albums.id) DESC 
            LIMIT 1
        """)
```

### Exiba os 5 clientes que mais fizeram pedidos. Qual é o valor da maior compra?

```python
consulta1("""
            SELECT 
                customers.first_name ||' '|| customers.last_name as full_name, 
                COUNT(invoices.total) as qtde_pedidos, 
                SUM(invoices.total) as total_pedidos
            FROM 
                customers
            JOIN 
                invoices
                    ON customers.id = invoices.customer_id
            GROUP BY 
                full_name
            ORDER BY 
                total_pedidos DESC 
            LIMIT 5
        """)

# Usei como criterio de desempate o valor total da compra
```

### Quais os emails dos 10 clientes que mais gastaram na loja? 

```python
consulta1("""
            SELECT 
                customers.first_name ||' '|| customers.last_name as full_name, 
                COUNT(invoices.total) as qtde_pedidos, 
                SUM(invoices.total) as total_pedidos,
                customers.email
            FROM 
                customers
            JOIN 
                invoices
                    ON customers.id = invoices.customer_id
            GROUP BY 
                full_name
            ORDER BY 
                total_pedidos DESC 
            LIMIT 10
        """)
```

### Qual o valor médio de cada compra? 


```python
consulta1("""
            SELECT
                customers.first_name ||' '|| customers.last_name as full_name,
                AVG(invoices.total) media_compras
            FROM customers
                JOIN invoices
                    ON customers.id = invoices.customer_id                    
            GROUP BY 
                full_name
            ORDER BY 
                media_compras DESC                                          
            """) 
```

### Quantas compras estão acima da média?

```python
#Media geral 
consulta1("""
            SELECT                     
               AVG(invoices.total) media_compras
            FROM customers
               JOIN invoices
                  ON customers.id = invoices.customer_id 
         """) 
```

```python
consulta1("""
            SELECT
                customers.first_name ||' '|| customers.last_name as full_name,
                invoices.total compras    
            FROM customers
                JOIN invoices
                    ON customers.id = invoices.customer_id 
            WHERE compras > (SELECT AVG(invoices.total) media_compras
                FROM customers
                JOIN invoices
                ON customers.id= invoices.customer_id)
            ORDER BY 
                compras DESC                                       
            """) 
```

### Vamos fazer uma playlist com as 12 músicas mais vendidas! 
Quais foram elas? Quais os generos e os artistas?

```python
consulta1("""
            SELECT
                COUNT(invoice_lines.track_id) nr_vendas,
                tracks.name music,  
                genres.name genres,
                tracks.composer
            FROM 
                invoice_lines
                    JOIN tracks
                        ON invoice_lines.track_id = tracks.id
                    JOIN genres
                        ON tracks.genre_id = genres.id
            GROUP BY 
                tracks.name
            ORDER BY 
                nr_vendas DESC 
            LIMIT 12
        """) 


```
