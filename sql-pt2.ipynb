```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('data/restaurant_sqlite.db')
def consulta(query):
    return pd.read_sql(query, conn)
```

```python
#Categorias dos Preços:
# < 6 - Barato
# >=6 e < 8 - Normal
# >= 8 - Caro 

consulta("""
         SELECT 
         *,
         CASE
            WHEN price < 6 THEN 'Barato' 
            WHEN price < 8 THEN 'Normal'
            ELSE 'Caro'
         END as categoria
         FROM dishes
         """)
```

```python
consulta("""
         SELECT 
         COUNT(price) AS qtd,
         CASE
            WHEN price < 6 THEN 'Barato' 
            WHEN price < 8 THEN 'Normal'
            ELSE 'Caro'
         END as categoria
         FROM dishes   
         GROUP BY categoria
         """)
```

```python
consulta("""
         SELECT 
         *,
         CASE
            WHEN price < 6 THEN 'Barato' 
            WHEN price < 8 THEN 'Normal'
            ELSE 'Caro'
         END as categoria
         FROM dishes
         ORDER BY categoria, price
         """)
```

```python
consulta("""
         SELECT 
         *,
         CASE
            WHEN price < 6 THEN 'Barato' 
            WHEN price < 8 THEN 'Normal'
            ELSE 'Caro'
         END as categoria
         FROM dishes
         WHERE categoria = 'Caro'
        """)
```

```python
consulta("""
         SELECT 
         c.firstname, 
         SUM(d.price) as sum_price
         FROM customers c
            INNER JOIN orders o 
            ON c.customerid = o.customerid
            INNER JOIN ordersdishes od
            ON o.orderid = od.orderid
            INNER JOIN dishes d
            ON od.dishid = d.dishid
            GROUP BY c.customerid
         """)
```

```python
consulta("""
         SELECT 
         c.firstname,
         c.lastname,
         SUM(d.price) as sum_price
         FROM customers c
            INNER JOIN orders o 
            ON c.customerid = o.customerid
            INNER JOIN ordersdishes od
            ON o.orderid = od.orderid
            INNER JOIN dishes d
            ON od.dishid = d.dishid
         GROUP BY c.customerid
         ORDER BY sum_price DESC
         LIMIT 5
         """) 
```
