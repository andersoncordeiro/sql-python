Vamos fazer uma playlist com as 12 músicas mais vendidas! 

Quais foram elas? Quais os generos e os artistas?



```python
consultar2('''SELECT playlist_tracks.track_id,tracks.name,invoice_lines.quantity,
COUNT(invoice_lines.quantity) as total_unidades, tracks.unit_price, 
artists.name,genres.name

FROM playlists, playlist_tracks, tracks, invoice_lines,albums,artists, genres,invoices
 where playlists.id = playlist_tracks.playlist_id
 and tracks.id = playlist_tracks.track_id 
 and invoice_lines.track_id = tracks.id
 and invoice_lines.invoice_id = invoices.id
 and tracks.album_id = albums.id
 and albums.artist_id = artists.id
 and genres.id = tracks.genre_id
 group by playlist_tracks.track_id
 ORDER BY total_unidades desc
 limit 12''')
```
