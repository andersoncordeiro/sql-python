# UTILIZANDO A BASE movies

#
Crie um dicionÃ¡rio de dados.

```python
dados = {}
cursor = movie.cursor()
names =[]
var_names = []
var_type = []
check_null =[]
#Master listing of all database objects in the database and the SQL used to create each object.
cursor.execute('SELECT name FROM sqlite_master WHERE TYPE = "table"')
name = cursor.fetchall()
for n in name: # [("Directors",), ("Actors",) ..]
    nome = n[0]
    cursor.execute(f"PRAGMA table_info({n[0]})") #[(0,id,integral,0,0,1), (...), (...)]
    for row in cursor.fetchall():
        names.append(n[0])
        var_names.append(row[1])
        var_type.append(row[2])
        #SELECT COUNT(*) FROM your_table WHERE your_column IS NULL;
        cursor.execute(f'SELECT COUNT(*) FROM {n[0]} WHERE {row[1]} IS NULL;')
        check = cursor.fetchall() # returns list of tuples ex.[(0,)]
        if check[0][0] >0:
            check_null.append("Yes")
        else:
            check_null.append("No")

dados = {"Table" : names, "Variable Names": var_names , "Variables Types" : var_type, "Null Values": check_null}       

df_movie = pd.DataFrame(dados)
```

#
Vamos fazer uma playlist com as 12 mÃºsicas mais vendidas! 
Quais foram elas? Quais os generos e os artistas?

```python
consulta_music('''SELECT 
                    t.name AS music,
                    g.name AS genre,
                    artists.name AS artist,
                    COUNT(il.track_id) total_sold
                FROM tracks t
                JOIN invoice_lines il
                    ON t.id = il.track_id
                JOIN invoices i
                    ON il.invoice_id = i.id
                JOIN genres g
                    ON t.genre_id = g.id
                JOIN albums a
                    ON t.album_id = a.id
                JOIN artists
                    ON a.artist_id = artists.id
                GROUP BY t.id
                ORDER BY total_sold DESC
                LIMIT 12
                ''')
#DÃºvidas
```

```python
consulta_music('''SELECT
                    t.id,
                    COUNT(il.track_id) total_sold
                FROM tracks t
                JOIN invoice_lines il
                    ON t.id = il.track_id
                JOIN invoices i
                    ON il.invoice_id = i.id
                JOIN genres g
                    ON t.genre_id = g.id
                JOIN albums a
                    ON t.album_id = a.id
                JOIN artists
                    ON a.artist_id = artists.id
                GROUP BY t.id
                ORDER BY total_sold DESC
                ''')
```
