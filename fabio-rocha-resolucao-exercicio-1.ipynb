```python
# Quantas compras estão acima da média? - 179
consulta("""
SELECT SUM(compras_acima_da_media) FROM
(SELECT invoices.id, invoices.total,
COUNT(CASE WHEN invoices.total > (SELECT (AVG(invoices.total)) FROM invoices JOIN customers ON customers.id = invoices.customer_id)
     THEN 1  
        ELSE Null END) as compras_acima_da_media
FROM invoices JOIN customers ON customers.id = invoices.customer_id GROUP BY invoices.id)
""")
```
