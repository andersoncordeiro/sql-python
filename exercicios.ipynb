```python
import sqlite3
import pandas as pd

conn_music_store = sqlite3.connect('data/music-store.db')

def consulta_music_store(query):
    return pd.read_sql(query, conn_music_store)
```

```python
#Qual artista possui mais albuns? 

consulta_music_store("""
                       SELECT artists.name, COUNT(albums.artist_id) FROM albums JOIN artists ON albums.artist_id = artists.id GROUP BY artists.name ORDER BY count(albums.artist_id) DESC LIMIT 1
                       """)

```

```python
#Quantos clientes são brasileiros?

consulta_music_store("""
                     SELECT country, COUNT(country)  FROM customers WHERE country='Brazil'""")
```

```python
# Qual genero possui mais músicas?

consulta_music_store("""
                     SELECT genres.name, count(tracks.genre_id) qtd FROM tracks JOIN genres ON tracks.genre_id = genres.id  GROUP BY genres.name ORDER BY count(tracks.genre_id) DESC LIMIT 1
                     """)
```

```python
# Qual genero possui mais albuns?

consulta_music_store("""
                     SELECT
                     genres.name,
                     tracks.genre_id qtd,
                     count(albums.id)
                     FROM tracks
                        JOIN genres 
                        ON tracks.genre_id = genres.id
                        JOIN albums 
                        ON tracks.album_id = albums.id
                    GROUP BY genres.name 
                    ORDER BY count(albums.id) DESC 
                    LIMIT 1
                     """)
```

```python
# Exiba os 5 clientes que mais fizeram pedidos. Qual é o valor da maior compra?
# como haviam muitos com 7 compras usei outro critério de desenpate pelo somatório dos pedidos.

consulta_music_store("""
                     SELECT
                     customers.first_name ||' '|| customers.last_name as full_name,
                     count(invoices.total) qtd_compras,
                     SUM(invoices.total) total
                     FROM customers
                        JOIN invoices
                        ON customers.id = invoices.customer_id
                     GROUP BY  full_name
                     ORDER BY total DESC
                     LIMIT 5
                     """)
```

```python
# Quais os emails dos 10 clientes que mais gastaram na loja? 

consulta_music_store("""
                     SELECT 
                     customers.first_name ||' '|| customers.last_name as full_name,
                     SUM(invoices.total) total,
                     customers.email
                     FROM customers
                        JOIN invoices
                        ON customers.id = invoices.customer_id
                     GROUP BY  full_name
                     ORDER BY total DESC
                     LIMIT 10
                     """)
```

```python
# Qual o valor médio de cada compra? 

consulta_music_store("""
                     SELECT
                     customers.first_name ||' '|| customers.last_name as full_name,
                     AVG(invoices.total)  media_compras
                     FROM customers
                        JOIN invoices
                        ON customers.id = invoices.customer_id                    
                     GROUP BY  full_name
                     ORDER BY media_compras                                           
                     """)
```

```python
# Quantas compras estão acima da média?


consulta_music_store("""
                     SELECT
                     customers.first_name ||' '|| customers.last_name as full_name,
                     invoices.total compras                                                       
                     
                     FROM customers
                        JOIN invoices
                        ON customers.id = invoices.customer_id 
                                                  
                     WHERE compras > (SELECT
                                       AVG(invoices.total)  media_compras
                                       FROM customers
                                          JOIN invoices
                                          ON customers.id = invoices.customer_id )   
                     ORDER BY compras DESC                                                                                 
                     """)
```

```python
#Vamos fazer uma playlist com as 12 músicas mais vendidas! 
#Quais foram elas? Quais os generos e os artistas?

consulta_music_store("""SELECT
                     count(invoice_lines.track_id) n_vendas,
                     tracks.name name_music,                     
                     genres.name name_genres,
                     tracks.composer artists
                     FROM 
                     invoice_lines
                        JOIN tracks
                        ON invoice_lines.track_id = tracks.id
                        JOIN genres
                        ON tracks.genre_id = genres.id                        
                     GROUP BY tracks.name
                     ORDER BY n_vendas DESC
                     LIMIT 12
                     """)
```

```python

#! PAIS QUE MAIS COMPROU - MAIS RICO. 
consulta_music_store("""
                     SELECT
                     billing_country,
                     sum(total) total
                     
                     FROM
                     invoices
                     GROUP BY billing_country
                     ORDER BY total DESC
                     LIMIT 1
                     """)
```

```python

#! QUANTO O MAIS RICO COMPROU DE CADA ALBUM
# como le comprou pouco de 1.99, e tem capital, é uma boa oferecer o produto mais caro 

consulta_music_store("""
                     SELECT
                     
                     invoices.billing_country,                                    
                     invoice_lines.unit_price preco,
                     count(invoice_lines.unit_price)
                     
                     FROM 
                     invoices
                     
                     JOIN invoice_lines
                     ON invoices.id = invoice_lines.invoice_id
                     WHERE invoices.billing_country = 'USA'
                     
                    GROUP BY preco
                    
                     """)
```

```python

#! PAIS QUE MENOS COMPROU POREM BELGICA NAO É POBRE, EMPATADO TEMOS A ARGENTINA QUE É POBRE  

consulta_music_store("""
                     SELECT
                     billing_country,
                     sum(total) total
                     
                     FROM
                     invoices
                     GROUP BY billing_country
                     ORDER BY total 
                     LIMIT 7
                     """)
```

```python

#! QUANTO O MAIS POBRE  COMPROU DE CADA ALBUM - usei o IDH como critério de riqueza.

##! Inicialmente indiquei a Argentina como país mais pobre, porém, possa ser que a Bélgica ou outro como
##! Austrália ou outro do bloco eutopeu tenha mais dinheiro para para gastar. 

consulta_music_store("""
                     SELECT
                     
                     invoices.billing_country,                    
                     invoice_lines.unit_price preco,
                     count(invoice_lines.unit_price)
                     
                     FROM 
                     invoices
                     
                     JOIN invoice_lines
                     ON invoices.id = invoice_lines.invoice_id
                     WHERE invoices.billing_country = 'Argentina'
                     
                    GROUP BY preco
                    
                     """)
```
